```python
plt.figure(figsize=(12,6),dpi=80)
data.groupby('Genre')['No_of_Votes'].mean().sort_values(ascending=False).head(15).plot(kind='bar',color='blue')
plt.xlabel('Genre',size=10)
plt.ylabel('Votes',size=10)
plt.title('Top 15 preferred votes by year',size=10)
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=80)
data.groupby('Director')['No_of_Votes'].mean().sort_values(ascending=False).head(15).plot(kind='line',color='orange')
plt.xlabel('DIrector',size=10)
plt.ylabel('Votes',size=10)
plt.title('Top 15 preferred votes by year',size=10)
plt.show()
```

```python
data['Gross'] = data['Gross'].fillna(0)
data.pivot_table(values=['Gross','No_of_Votes','Meta_score'], index='Released_Year', aggfunc='mean').sort_values(by=['Meta_score'])
```
